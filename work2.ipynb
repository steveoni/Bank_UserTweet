## Group data
After the data have being classified in to different categorise of sentiment via the javascript user interface, they are now concatenated togeter to form a single file 

```python
import pandas as pd
import os
```

```python
from os import listdir
from os.path import isfile, join

mypath= "clean/bank_data/"
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
onlyfiles
```

```python
df = pd.concat([pd.read_csv("clean/bank_data/"+csv,lineterminator='\n') for csv in onlyfiles],axis=0)
```
